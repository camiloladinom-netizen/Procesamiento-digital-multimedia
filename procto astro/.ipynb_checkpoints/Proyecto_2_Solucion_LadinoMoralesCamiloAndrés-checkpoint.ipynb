## 3. Filtrado y denoising

**Tarea:** implementa y compara al menos tres métodos de filtrado. Ejemplos: filtro promedio, filtro gaussiano, filtro de mediana, filtro bilateral o denoising con wavelets.

Discute cuál método preserva mejor los objetos débiles y cuál elimina mejor los píxeles calientes.

````python
# Preparar imagen de trabajo convirtiendo a flotante y normalizando a [0, 1]
example_img = test_img.astype(np.float32)  # Convierte la imagen de ejemplo a flotante
img_norm = (example_img - example_img.min()) / (example_img.max() - example_img.min() + 1e-6) # Normaliza para evitar division por cero
img = img_norm  # Alias para usar un nombre mas corto en el resto del codigo

# Filtro gaussiano para suavizar ruido manteniendo estructuras grandes
sigma_gauss = 1.0  # Desviacion estandar del filtro gaussiano en pixeles
gaussian_filtered = filters.gaussian(img, sigma=sigma_gauss, preserve_range=True)  # Aplica filtro gaussiano

# Filtro de mediana para eliminar pixeles calientes y ruido impulsivo
radius = 1  # Radio del elemento estructurante para la mediana (en pixeles)
selem = morphology.disk(radius)  # Crea un disco como elemento estructurante para la mediana
median_filtered = filters.median(img, footprint=selem)  # Aplica filtro de mediana sobre la imagen

# Denoising basado en wavelets (pywt) como tercer metodo
wavelet = 'db2'  # Tipo de wavelet discreta a utilizar (Daubechies 2)
level = 2  # Numero de niveles de descomposicion para la transformada wavelet 2D
coeffs = pywt.wavedec2(img, wavelet=wavelet, level=level)  # Calcula coeficientes wavelet 2D
cA, cD_list = coeffs[0], coeffs[1:]  # Separa coeficientes de aproximacion y detalles

# Estimar sigma del ruido usando coeficientes de detalle del primer nivel
sigma_est = np.median(np.abs(cD_list[0][0])) / 0.6745  # Estimacion robusta de sigma del ruido
threshold = sigma_est * np.sqrt(2 * np.log(img.size))  # Umbral universal segun Donoho-Johnstone

# Aplicar umbral suave (soft-thresholding) a todos los coeficientes de detalle
cD_list_thresh = []  # Lista para guardar detalles umbralizados
for (cH, cV, cD) in cD_list:  # Recorre los niveles de detalle horizontales, verticales y diagonales
    cH_t = pywt.threshold(cH, threshold, mode='soft')  # Aplica umbral suave a coeficientes horizontales
    cV_t = pywt.threshold(cV, threshold, mode='soft')  # Aplica umbral suave a coeficientes verticales
    cD_t = pywt.threshold(cD, threshold, mode='soft')  # Aplica umbral suave a coeficientes diagonales
    cD_list_thresh.append((cH_t, cV_t, cD_t))  # Agrega el trio umbralizado a la lista

coeffs_thresh = [cA] + cD_list_thresh  # Reconstruye lista de coeficientes con detalles filtrados
wavelet_denoised = pywt.waverec2(coeffs_thresh, wavelet=wavelet)  # Reconstruye imagen denoised
wavelet_denoised = np.clip(wavelet_denoised, 0.0, 1.0)  # Limita valores al rango [0, 1]

# Visualizar imagen original y las versiones filtradas
fig, axes = plt.subplots(1, 4, figsize=(16, 4))  # Crea figura con cuatro subplots en una fila
axes[0].imshow(img, cmap='gray')  # Muestra la imagen original normalizada
axes[0].set_title('Original')  # Titulo de la imagen original
axes[0].axis('off')  # Oculta ejes para resaltarla

axes[1].imshow(gaussian_filtered, cmap='gray')  # Muestra imagen filtrada con gaussiano
axes[1].set_title('Gaussiano')  # Titulo para el filtro gaussiano
axes[1].axis('off')  # Oculta ejes

axes[2].imshow(median_filtered, cmap='gray')  # Muestra imagen filtrada con mediana
axes[2].set_title('Mediana')  # Titulo para el filtro de mediana
axes[2].axis('off')  # Oculta ejes

axes[3].imshow(wavelet_denoised, cmap='gray')  # Muestra imagen denoised por wavelets
axes[3].set_title('Wavelet')  # Titulo para el metodo wavelet
axes[3].axis('off')  # Oculta ejes

plt.tight_layout()  # Ajusta espacios entre subplots
plt.show()  # Muestra la figura con las cuatro imagenes
````

## 4. Estimación y corrección de fondo

El fondo astronómico no es constante. Debes estimar una imagen de fondo suave $\hat{B}(x,y)$ y restarla:

$$I_{\mathrm{corr}}(x,y)=I_{\mathrm{obs}}(x,y)-\hat{B}(x,y)$$

**Tarea:** prueba al menos un método de corrección de fondo.

````python
# Seleccionar una imagen denoised (mediana) para estimar el fondo
denoised_img = median_filtered  # Usa la imagen filtrada con mediana como base para el fondo

# Estimar fondo con un filtro gaussiano de gran sigma (fondo suave)
sigma_bg = 10.0  # Desviacion estandar grande para capturar solo variaciones suaves del fondo
background_est = filters.gaussian(denoised_img, sigma=sigma_bg, preserve_range=True)  # Imagen de fondo estimado

# Corregir la imagen restando el fondo estimado
corrected_img = denoised_img - background_est  # Resta el fondo de la imagen denoised
corrected_img = corrected_img - corrected_img.min()  # Desplaza para que la minima intensidad sea cero
corrected_img = corrected_img / (corrected_img.max() + 1e-6)  # Normaliza a rango [0, 1] para estabilidad

# Visualizar imagen original normalizada, fondo estimado e imagen corregida
fig, axes = plt.subplots(1, 3, figsize=(15, 4))  # Crea figura con tres subplots en una fila
axes[0].imshow(img, cmap='gray')  # Muestra imagen original normalizada
axes[0].set_title('Original normalizada')  # Titulo para imagen original
axes[0].axis('off')  # Oculta ejes

axes[1].imshow(background_est, cmap='gray')  # Muestra el fondo estimado
axes[1].set_title('Fondo estimado')  # Titulo para la imagen de fondo
axes[1].axis('off')  # Oculta ejes

axes[2].imshow(corrected_img, cmap='gray')  # Muestra la imagen corregida
axes[2].set_title('Imagen corregida')  # Titulo para imagen corregida
axes[2].axis('off')  # Oculta ejes

plt.tight_layout()  # Ajusta espacios entre los subplots
plt.show()  # Muestra la figura con las tres imagenes
````

## 5. Realce de contraste

**Tarea:** aplica normalización, transformación gamma, ecualización de histograma o ecualización adaptativa. Explica cómo cambia la visibilidad de los objetos débiles.

````python
# Aplicar realce de contraste mediante ecualizacion adaptativa (CLAHE)
clip_limit = 0.01  # Limite de recorte para controlar el contraste local en CLAHE
corrected_clahe = exposure.equalize_adapthist(corrected_img, clip_limit=clip_limit)  # Aplica CLAHE sobre imagen corregida

# Comparar imagen corregida antes y despues del realce de contraste
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # Crea figura con dos subplots
axes[0].imshow(corrected_img, cmap='gray')  # Muestra imagen corregida sin realce
axes[0].set_title('Corregida sin realce')  # Titulo para imagen sin CLAHE
axes[0].axis('off')  # Oculta ejes

axes[1].imshow(corrected_clahe, cmap='gray')  # Muestra imagen con contraste realzado
axes[1].set_title('Corregida con CLAHE')  # Titulo para imagen con CLAHE
axes[1].axis('off')  # Oculta ejes

plt.tight_layout()  # Ajusta espacios entre subplots
plt.show()  # Muestra la figura comparativa
````

## 6. Segmentación

Debes construir una máscara binaria:

$$M(x,y)=1 \quad \text{si hay objeto detectado}$$

**Tarea:** implementa un método de umbralización. Puedes usar umbral global, Otsu, adaptativo, o un umbral de la forma:

$$T = \mu_B + k\sigma_B$$

````python
# Seleccionar imagen para segmentacion (corregida y con contraste realzado)
seg_img = corrected_clahe  # Usa la imagen con contraste realzado como base para segmentar

# Estimar estadisticas del fondo a partir de pixeles de baja intensidad
background_mask = seg_img < 0.3  # Considera fondo los pixeles con intensidad relativamente baja
background_values = seg_img[background_mask]  # Extrae intensidades de esos pixeles
mu_B = background_values.mean()  # Media de intensidades del fondo
sigma_B = background_values.std()  # Desviacion estandar del fondo

# Definir umbral estadistico T = mu_B + k * sigma_B
k = 3.0  # Factor multiplicador sobre la desviacion estandar del fondo
T = mu_B + k * sigma_B  # Umbral estadistico basado en la media y desviacion del fondo

# Calcular tambien el umbral global de Otsu como referencia
otsu_T = filters.threshold_otsu(seg_img)  # Umbral global de Otsu calculado sobre la imagen

print(f'Umbral T = mu_B + k*sigma_B = {T:.4f}')  # Imprime valor del umbral estadistico
print(f'Umbral de Otsu = {otsu_T:.4f}')  # Imprime valor del umbral de Otsu

# Elegir un umbral final (por ejemplo, el mayor para ser mas conservador)
final_T = max(T, otsu_T)  # Toma el umbral mas alto entre el estadistico y Otsu

# Generar mascara binaria aplicando el umbral final
binary_mask = seg_img > final_T  # Marca como 1 los pixeles por encima del umbral

# Visualizar imagen de entrada y mascara resultante
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # Crea figura con dos subplots
axes[0].imshow(seg_img, cmap='gray')  # Muestra la imagen utilizada para segmentacion
axes[0].set_title('Imagen para segmentacion')  # Titulo descriptivo
axes[0].axis('off')  # Oculta ejes

axes[1].imshow(binary_mask, cmap='gray')  # Muestra la mascara binaria obtenida
axes[1].set_title('Mascara binaria')  # Titulo para la mascara
axes[1].axis('off')  # Oculta ejes

plt.tight_layout()  # Ajusta espacios entre subplots
plt.show()  # Muestra figura con imagen y mascara
````

## 7. Morfología matematica

**Tarea:** limpia la máscara usando erosión, dilatación, apertura, cierre, relleno de huecos o eliminación de objetos pequeños.

````python
# Aplicar operaciones morfologicas para limpiar la mascara binaria
selem_morph = morphology.disk(1)  # Elemento estructurante disco de radio 1 pixel
opened_mask = morphology.opening(binary_mask, selem_morph)  # Apertura: elimina ruido pequeño aislado
closed_mask = morphology.closing(opened_mask, selem_morph)  # Cierre: rellena huecos pequenos en los objetos

min_area = 5  # Area minima en pixeles para considerar un objeto valido
clean_mask = morphology.remove_small_objects(closed_mask, max_size=min_area)  # Elimina regiones pequenas

# Visualizar mascaras antes y despues de la limpieza morfologica
fig, axes = plt.subplots(1, 3, figsize=(15, 4))  # Crea figura con tres subplots
axes[0].imshow(binary_mask, cmap='gray')  # Muestra mascara binaria inicial
axes[0].set_title('Mascara inicial')  # Titulo para mascara original
axes[0].axis('off')  # Oculta ejes

axes[1].imshow(opened_mask, cmap='gray')  # Muestra mascara despues de la apertura
axes[1].set_title('Tras apertura')  # Titulo para mascara abierta
axes[1].axis('off')  # Oculta ejes

axes[2].imshow(clean_mask, cmap='gray')  # Muestra mascara final limpia
axes[2].set_title('Mascara limpia')  # Titulo para mascara ya filtrada
axes[2].axis('off')  # Oculta ejes

plt.tight_layout()  # Ajusta espacios entre los tres subplots
plt.show()  # Muestra la figura con las mascaras
````

## 8. Componentes conectados y medición de objetos

**Tarea:** etiqueta los componentes conectados de la máscara final. Para cada objeto detectado estima:

- centroide $(\hat{x},\hat{y})$,
- area,
- brillo total,
- brillo medio.

El resultado puede quedar como un `DataFrame` interno llamado `pred_catalog`.

````python
# Definir funcion que implementa el pipeline de deteccion en una sola imagen
def detect_sources_in_image(img_path):  # Funcion que procesa una imagen y devuelve un DataFrame de objetos
    img = io.imread(img_path).astype(np.float32)  # Lee la imagen desde disco y la convierte a flotante
    img_norm = (img - img.min()) / (img.max() - img.min() + 1e-6)  # Normaliza intensidades al rango [0, 1]

    # Paso 1: denoising con filtro de mediana
    selem = morphology.disk(1)  # Elemento estructurante pequeno para la mediana
    denoised = filters.median(img_norm, footprint=selem)  # Aplica filtro de mediana

    # Paso 2: estimacion de fondo con filtro gaussiano ancho
    sigma_bg = 10.0  # Sigma grande para capturar solo variaciones suaves del fondo
    background = filters.gaussian(denoised, sigma=sigma_bg, preserve_range=True)  # Estima fondo

    # Paso 3: correccion de fondo
    corrected = denoised - background  # Resta el fondo estimado
    corrected = corrected - corrected.min()  # Desplaza para que el minimo sea cero
    corrected = corrected / (corrected.max() + 1e-6)  # Normaliza al rango [0, 1]

    # Paso 4: realce de contraste con CLAHE
    clip_limit = 0.01  # Limite de recorte para controlar contraste local
    enhanced = exposure.equalize_adapthist(corrected, clip_limit=clip_limit)  # Aplica CLAHE

    # Paso 5: estadisticas del fondo para construir umbral
    background_mask = enhanced < 0.3  # Considera fondo los pixeles de baja intensidad
    background_values = enhanced[background_mask]  # Extrae intensidades de fondo
    mu_B = background_values.mean()  # Media del fondo
    sigma_B = background_values.std()  # Desviacion estandar del fondo
    k = 3.0  # Factor para definir umbral estadistico
    T = mu_B + k * sigma_B  # Umbral estadistico basado en fondo
    otsu_T = filters.threshold_otsu(enhanced)  # Umbral de Otsu calculado sobre la imagen
    final_T = max(T, otsu_T)  # Elige el umbral mas conservador (mas alto)

    # Paso 6: segmentacion binaria
    binary = enhanced > final_T  # Mascara de pixeles considerados como objetos

    # Paso 7: limpieza morfologica
    selem_morph = morphology.disk(1)  # Elemento estructurante para apertura y cierre
    opened = morphology.opening(binary, selem_morph)  # Apertura para eliminar ruido pequeno
    closed = morphology.closing(opened, selem_morph)  # Cierre para rellenar huecos pequenos
    min_area = 5  # Area minima para conservar un objeto
    clean = morphology.remove_small_objects(closed, max_size=min_area)  # Elimina componentes pequenos

    # Paso 8: etiquetado de componentes conectados y medicion
    labeled = measure.label(clean)  # Etiqueta componentes conectados en la mascara limpia
    props = measure.regionprops(labeled, intensity_image=img_norm)  # Calcula propiedades por region

    rows = []  # Lista donde se acumularan las filas del catalogo predicho
    for region in props:  # Recorre cada objeto detectado en la imagen
        y_centroid, x_centroid = region.centroid  # Obtiene centroide (fila, columna) de la region
        area = region.area  # Area del objeto en pixeles
        total_brightness = region.image_intensity.sum()  # Brillo total usando intensidades normalizadas
        mean_brightness = region.intensity_mean  # Brillo medio de la region

        rows.append({  # Agrega un diccionario con la informacion del objeto detectado
            'filename': img_path.name,  # Nombre del archivo de la imagen
            'object_id_pred': region.label,  # Identificador del objeto predicho
            'x_pred': x_centroid,  # Coordenada x del centroide (columna)
            'y_pred': y_centroid,  # Coordenada y del centroide (fila)
            'area': area,  # Area en pixeles del objeto
            'total_brightness': total_brightness,  # Brillo total del objeto
            'mean_brightness': mean_brightness,  # Brillo medio del objeto
        })  # Cierra el diccionario del objeto

    pred_df = pd.DataFrame(rows)  # Construye un DataFrame a partir de la lista de objetos
    return pred_df  # Devuelve el catalogo predicho para esta imagen

# Aplicar el pipeline a todas las imagenes del dataset
pred_catalog_list = []  # Lista para acumular catalogos predichos por imagen
for img_path in image_files:  # Recorre todas las imagenes disponibles
    pred_df = detect_sources_in_image(img_path)  # Detecta objetos en la imagen actual
    pred_catalog_list.append(pred_df)  # Agrega el DataFrame de esta imagen a la lista

# Unir todos los resultados en un unico DataFrame global
pred_catalog = pd.concat(pred_catalog_list, ignore_index=True)  # Une resultados de todas las imagenes
pred_catalog.head()  # Muestra las primeras filas como verificacion rapida
````

## 9. Emparejamiento con el catálogo verdadero

Para calcular el error de centroide, cada detección debe emparejarse con un objeto verdadero cercano. Define una tolerancia $r_{tol}$ en píxeles.

Una detección puede considerarse correcta si:

$$\sqrt{(x_i-\hat{x}_j)^2+(y_i-\hat{y}_j)^2} < r_{tol}$$

````python
# Definir tolerancia espacial r_tol para emparejar centroides entre verdad y prediccion
r_tol = 3.0  # Tolerancia en pixeles para considerar una deteccion como correcta

# Asegurar nombres de columnas estandar en el catalogo verdadero
true_catalog_renamed = true_catalog.rename(columns={'x': 'x_true', 'y': 'y_true'})  # Renombra columnas si es necesario

matched_rows = []  # Lista para guardar emparejamientos correctos
unmatched_true = []  # Lista para objetos verdaderos sin deteccion
unmatched_pred = []  # Lista para detecciones sin objeto verdadero asociado

# Recorrer cada imagen por separado para emparejar objetos
for img_path in image_files:  # Recorre todas las imagenes del dataset
    fname = img_path.name  # Obtiene el nombre de archivo de la imagen

    true_img = true_catalog_renamed[true_catalog_renamed['filename'] == fname]  # Filtra objetos verdaderos de esta imagen
    pred_img = pred_catalog[pred_catalog['filename'] == fname]  # Filtra objetos predichos de esta imagen

    used_pred_indices = set()  # Conjunto para marcar detecciones ya emparejadas

    for idx_true, row_true in true_img.iterrows():  # Recorre cada objeto verdadero en la imagen
        x_t = row_true['x_true']  # Coordenada x verdadera
        y_t = row_true['y_true']  # Coordenada y verdadera

        # Calcular distancia euclidiana a todos los centroides predichos
        dx = pred_img['x_pred'] - x_t  # Diferencia en x entre predicho y verdadero
        dy = pred_img['y_pred'] - y_t  # Diferencia en y entre predicho y verdadero
        distances = np.sqrt(dx**2 + dy**2)  # Distancia euclidiana para cada deteccion

        if len(distances) == 0:  # Caso en que no hay detecciones en esta imagen
            unmatched_true.append((fname, x_t, y_t))  # Registra objeto verdadero sin match
            continue  # Pasa al siguiente objeto verdadero

        best_idx = distances.idxmin()  # Indice del objeto predicho mas cercano
        best_dist = distances.loc[best_idx]  # Distancia minima encontrada

        if best_dist < r_tol and best_idx not in used_pred_indices:  # Verifica tolerancia y no reutilizacion
            used_pred_indices.add(best_idx)  # Marca esta deteccion como utilizada
            row_pred = pred_img.loc[best_idx]  # Obtiene fila correspondiente del catalogo predicho

            matched_rows.append({  # Agrega un emparejamiento correcto a la lista
                'filename': fname,  # Nombre del archivo de la imagen
                'x_true': x_t,  # Coordenada x verdadera
                'y_true': y_t,  # Coordenada y verdadera
                'x_pred': row_pred['x_pred'],  # Coordenada x predicha
                'y_pred': row_pred['y_pred'],  # Coordenada y predicha
                'distance': best_dist,  # Distancia entre los centroides
            })  # Cierra diccionario del emparejamiento
        else:  # Caso en que la deteccion mas cercana no cumple tolerancia o ya fue usada
            unmatched_true.append((fname, x_t, y_t))  # Registra objeto verdadero sin deteccion valida

    # Detectar predicciones que quedaron sin emparejar
    for idx_pred, row_pred in pred_img.iterrows():  # Recorre todas las detecciones de la imagen
        if idx_pred not in used_pred_indices:  # Verifica si alguna deteccion no se uso
            unmatched_pred.append((fname, row_pred['x_pred'], row_pred['y_pred']))  # Registra deteccion sin objeto verdadero

# Construir DataFrame con los emparejamientos obtenidos
matches_df = pd.DataFrame(matched_rows)  # Crea DataFrame a partir de la lista de matches
matches_df.head()  # Muestra las primeras filas como verificacion
````

## 10. Métricas obligatorias

Debes reportar claramente:

### Error de conteo

$$E_N = |N_{pred} - N_{true}|$$

Reporta media, mediana y/o histograma de $E_N$ sobre el conjunto de imágenes.

### Error de centroide

$$e_c^{(i)} = \sqrt{(x_i-\hat{x}_i)^2+(y_i-\hat{y}_i)^2}$$

Reporta media, mediana y desviacion estandar del error de centroide en píxeles.

````python
# Calcular error de conteo E_N por imagen
counts_true = true_catalog_renamed['filename'].value_counts().sort_index()  # Conteo de objetos verdaderos por imagen
counts_pred = pred_catalog['filename'].value_counts().sort_index()  # Conteo de objetos predichos por imagen

all_filenames = sorted(set(counts_true.index) | set(counts_pred.index))  # Union de todos los nombres de archivo

EN_list = []  # Lista para guardar errores de conteo por imagen
for fname in all_filenames:  # Recorre todas las imagenes
    N_true = counts_true.get(fname, 0)  # Numero de objetos verdaderos (0 si no hay)
    N_pred = counts_pred.get(fname, 0)  # Numero de objetos predichos (0 si no hay)
    E_N = abs(N_pred - N_true)  # Calcula el error absoluto de conteo
    EN_list.append({'filename': fname, 'N_true': N_true, 'N_pred': N_pred, 'E_N': E_N})  # Agrega fila a la lista

EN_df = pd.DataFrame(EN_list)  # Convierte la lista a DataFrame

print('Resumen de error de conteo E_N por imagen:')  # Encabezado del resumen de E_N
print(EN_df.head())  # Muestra las primeras filas del resumen de E_N

print('Estadisticas globales de E_N:')  # Encabezado para estadisticas globales
print(f'Media de E_N: {EN_df['E_N'].mean():.3f}')  # Imprime media de E_N sobre todas las imagenes
print(f'Mediana de E_N: {EN_df['E_N'].median():.3f}')  # Imprime mediana de E_N
print(f'Desviacion estandar de E_N: {EN_df['E_N'].std():.3f}')  # Imprime desviacion estandar de E_N

# Calcular error de centroide para objetos emparejados
matches_df['e_c'] = matches_df['distance']  # Copia la distancia como error de centroide e_c

print('Estadisticas del error de centroide e_c (en pixeles):')  # Encabezado para estadisticas de e_c
print(f'Media de e_c: {matches_df['e_c'].mean():.3f}')  # Media del error de centroide
print(f'Mediana de e_c: {matches_df['e_c'].median():.3f}')  # Mediana del error de centroide
print(f'Desviacion estandar de e_c: {matches_df['e_c'].std():.3f}')  # Desviacion estandar de e_c
````
